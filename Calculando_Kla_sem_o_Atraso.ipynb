CURVA 1

```
#Importanto biblioteca
import numpy as np
import pandas as pd

#Importando os dados experimentais
importado = pd.read_excel("medidaskla.xlsx")
#Renomeando os dados experimentais
dados_np = importado.values
t1 = dados_np[:,0]
c1 = dados_np[:,1]

#Linearizando: -ln(1-(CO2/C*O2)) = Kla*t
#Considerei a concentração na saturação como 100%
y1 = np.array(-np.log(1-(c1/100)))

# Regressão linear: y=a*x+b
mod_linear1 = np.polyfit(t1,y1,1)
a1 = (mod_linear1[0])
b1 = (mod_linear1[1])
y1_mod = a1*t1+b1

#Calculando o coeficiente de correlação R²
import sklearn.metrics
from sklearn.metrics import r2_score
r2 = r2_score(y1_mod,y1)

#Plotando o resultados 
import matplotlib.pyplot as plt
plt.plot(t1,y1,"oy", label= "Pontos experimentais linearizados")
plt.plot(t1,y1_mod,"-b", label = "Modelo")
plt.legend()
plt.xlabel("Tempo (s)")
plt.ylabel("-ln(1-(C02/C*02)")
plt.title("Ajuste linear - Método Dinâmico")
print ("O coeficiente angular é:", (a1), " = kla (s-¹)")
print ("O coeficiente de correlação (R²) é:",(r2))
```
